```python
import numpy as np

#Примеры создания массивов
# из списка
arr_list = np.array([1, 2, 3, 4])
# двумерный массив нулей
arr_zeros = np.zeros((2, 3))
# одномерный массив единиц
arr_ones = np.ones(5)
# [0, 2, 4, 6, 8]
arr_arange = np.arange(0, 10, 2) 

# Атрибуты массива
print("Форма arr_zeros:", arr_zeros.shape)
print("Тип данных arr_zeros:", arr_zeros.dtype)

# Индексирование и срезы
print("\nПервый элемент arr_list:", arr_list[0])
print("Срез arr_arange[1:3]:", arr_arange[1:3])

# Векторные операции
sum_arrays = arr_list + np.array([10, 20, 30, 40])
print("\nПоэлементное сложение:", sum_arrays)

mean_value = np.mean(arr_list)
print("Среднее значение arr_list:", mean_value)
```
